```python
# Set default catalog and schema
catalog = dbutils.widgets.get("catalog")
schema = dbutils.widgets.get("schema")
spark.sql(f"USE CATALOG {catalog}")
spark.sql(f"USE SCHEMA {schema}")
```

```python
sql_command_drop = f"""DROP TABLE IF EXISTS {catalog}.{schema}.people_with_type"""

sql_command_create = f"""CREATE OR REPLACE TABLE {catalog}.{schema}.people_with_type
COMMENT 'type column based on role'
AS
SELECT *,
  CASE WHEN user_id< '10' THEN 1 ELSE 0 END AS type
FROM {catalog}.{schema}.people
"""
```

```python
spark.sql(sql_command_drop)
spark.sql(sql_command_create)
```
